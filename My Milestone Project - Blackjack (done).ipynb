```python
# player bank account class

class Player:
    
    def __init__(self, account):
        self.account = account
        
    def __str__(self):
        return 'Player has $'+str(self.account)
    
    # method to add winnings
    def add_winning(self, amount):
        self.account += amount

    # method to deduct losings
    def deduct_loss(self, amount):
        self.account -= amount
            
```

```python
def player_start():
    start_or_not='?'
    while not(start_or_not=='Y' or start_or_not=='N'):
        start_or_not = input('Do you want to start the game? Use Y/N:').upper()
        if start_or_not=='Y' or start_or_not=='N':
            break                                               # break the while loop
        else:
            print('Please input only Y or N')
            
    return start_or_not
```

```python
# Global variables

# immutables
suits = ('Hearts', 'Diamonds', 'Spade', 'Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')

# convert rank to integer value for comparison
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 
          'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':[1,11]}
```

```python
# create deck of cards class
import random

class Deck:
    
    def __init__(self):
        # for deck creation
        self.card_deck = []   # card_Deck is parked under Deck class
    
        for suit in suits:
            for rank in ranks:
                card = [rank,suit]
                self.card_deck.append(card)
    
    # define shuffling method within Deck class
    def shuffle(self):
        random.shuffle(self.card_deck)  #random.shuffle will actually perform shuffle on self.card_deck and return not required 
        
    # define dealing method within Deck class
    def deal_once(self):
        deal_card = self.card_deck.pop(0)        
        return deal_card
    
```

```python
# define bet function

def bet():
    
    bet_amt = 0.0 
    
    while type(bet_amt) != int:
        
        bet_amt = input('How much do you want to bet? Please put in an integer: ')
        
        try:
            bet_amt = int(bet_amt)
                        
        except:
            print('Please put in an integer')
            continue       
        
    return bet_amt
        
```

```python
# create Player's cards class

class PlayerCards:
    
    def __init__(self):
        self.player_cards = []

    def __str__(self):
        return self.player_cards        
    
    
    # define method to add card to player's cards
    def add_card(self,deal_card):
        self.player_cards.append(deal_card)
      

    # define method to count points in player's cards
    def count_points(self):
        total_pts=0
        no_of_ace=0
        for card in self.player_cards:
            if card[0] != 'Ace':
                points = values[card[0]]
                total_pts += points 
            elif card[0] == 'Ace':
                no_of_ace += 1  # don't add the points of ace first since its value need to be assessed based on other cards points.
                                # take note of the no of Ace cards for final calculation
        
        if no_of_ace ==0:
            return total_pts
        
        elif no_of_ace>0:
            if (total_pts+11+(no_of_ace-1)*1) > 21:
                total_pts += ((no_of_ace)*1)
            else:
                total_pts += (11+(no_of_ace-1)*1)
            return total_pts    
```

```python
# create Computer's cards class

class ComputerCards:
    
    def __init__(self):
        self.computer_cards = []

    def __str__(self):
        return self.computer_cards        
    
    # define method to add card to computer's cards
    def add_card(self,deal_card):
        self.computer_cards.append(deal_card)
        
    # define method to count points in computer's cards
    def count_points(self):
        total_pts=0
        no_of_ace=0
        for card in self.computer_cards:
            if card[0] != 'Ace':
                points = values[card[0]]
                total_pts += points 
            elif card[0] == 'Ace':
                no_of_ace += 1  # don't add the points of ace first since its value need to be assessed based on other cards points.
                                # take note of the no of Ace cards for final calculation
        
        if no_of_ace ==0:
            return total_pts
        
        elif no_of_ace>0:
            if (total_pts+11+(no_of_ace-1)*1) > 21:
                total_pts += ((no_of_ace)*1)
            else:
                total_pts += (11+(no_of_ace-1)*1)
            return total_pts    
```

```python
# define function to ask player if he/she wants to hit or stay 

def hit_or_stay():
    
    player_hit = 2
    
    while (player_hit == 1 or player_hit == 0)==False:
        player_hit = input('Do you want to hit or stay? Input 1 for Hit, or 0 for Stay. ')
        
        try:
            player_hit = int(player_hit)
            
            if not(player_hit == 1 or player_hit == 0)==True:
                print('Please put in only 0 or 1') # if condition True, print and will then go back to start of while loop to assess 'player_hit' criteria
            else:
                return player_hit                  # else, return 'player_hit' which will also end the while loop
                
        
        except:         # if input string cannot pass the int() function, skip the error msg & continue on with the code block here
            print('Please put in only 0 or 1')
            continue  # go back to start of while loop
```

```python
def player_continue():
    continue_or_not='?'
    while not(continue_or_not=='Y' or continue_or_not=='N'):
        continue_or_not = input('Do you want to continue the game? Use Y/N:').upper()
        if continue_or_not=='Y' or continue_or_not=='N':
            break                                               # break the while loop
        else:
            print('Please input only Y or N')
            
    return continue_or_not
```

```python
#### MAIN BLACKJACK MAIN LOGIC:


### Setup player bank starting account = $1000
player_bank = Player(1000)

### Setup game loop
start_or_not = player_start()
if start_or_not == 'Y':
    game_on = True
else:
    game_on = False

while game_on==True:

    ## Reset player bust and computer bust indicators for this round (0 means not busted)
    player_bust = 0
    computer_bust = 0

    ## create deck of cards
    cards = Deck()

    ## shuffle cards
    cards.shuffle()

    ## deal 2 cards each - player's cards are shown, dealer's cards will have 1 hidden card
    computer_cards = ComputerCards()
    player_cards = PlayerCards()
    
    dealt_card = cards.deal_once()
    computer_cards.add_card(dealt_card)

    dealt_card = cards.deal_once()
    player_cards.add_card(dealt_card)

    dealt_card = cards.deal_once()
    computer_cards.add_card(dealt_card)

    dealt_card = cards.deal_once()
    player_cards.add_card(dealt_card)

    comp_revealed_str = 'Computer dealer cards are : [Unknown], '+str(computer_cards.computer_cards[1])
    comp_cards_str = 'Computer dealer cards are : '+str(computer_cards.computer_cards[0])+', '+str(computer_cards.computer_cards[1])
    player_revealed_str = 'Player cards are : '+str(player_cards.player_cards[0])+', '+str(player_cards.player_cards[1])

    print(comp_revealed_str)
    print(player_revealed_str)

    ## Ask player for bet
    bet_amount = bet()

    while bet_amount > player_bank.account:
        print('The bet amount is more than what you have, please reduce the amount to bet!')
        bet_amount = bet()

    ## Ask player to hit or stay, and assess if player bust

    while True:
        player_to_hit = hit_or_stay()

        if player_to_hit == 1:
            dealt_card = cards.deal_once()
            player_cards.add_card(dealt_card)
            print(len(cards.card_deck))           #TEST
            player_revealed_str = player_revealed_str+', '+str(player_cards.player_cards[-1])
            print(player_revealed_str)
            player_pts = player_cards.count_points()
            
            if player_pts > 21:
                print('Player has busted, you lost')
                player_bust = 1
                break               # break will only break out of the inner nested loop that it is inside of, which is the the ask-player-to-hit-or-stay decision while loop 
            else:
                continue

        elif player_to_hit == 0:
            print(player_revealed_str)
            player_pts = player_cards.count_points()
            print('Player total points is :'+str(player_pts))
            break                   # break out of the ask-player-to-hit-or-stay decision while loop only when player decides to stay


    ## if player did not bust, then computer dealer will hit or stay based on the algorithm programmed here, 
    ## and finally assess if computer bust 
    if player_bust == 0:
        
        # computer decision logic / strategy
        computer_pts = computer_cards.count_points()
        
        while (computer_pts < player_pts) or (computer_pts == player_pts and computer_pts<16):
            # computer hits
            print('Dealer hits')
            dealt_card = cards.deal_once()
            computer_cards.add_card(dealt_card)
            print(len(cards.card_deck))           #TEST
            comp_cards_str = comp_cards_str+', '+str(computer_cards.computer_cards[-1])
            comp_revealed_str = comp_revealed_str+', '+str(computer_cards.computer_cards[-1])
            print(comp_revealed_str)
            computer_pts = computer_cards.count_points()
                
        # computer stays
        print('Dealer stays')
        print(comp_revealed_str)        
         
        # computer reveal the hidden card and check whether computer dealer is busted
        print('Dealer reveals hidden card')
        print(comp_cards_str)
        if computer_pts>21:
            print('Dealer has busted. You win!')
            computer_bust = 1
            
    ## If both did not bust, compare to see who wins / lose
    if computer_bust == 0 and player_bust == 0:
        # print points from both sides
        print('Computer dealer total points is :'+str(computer_pts))
        print('Player total points is :'+str(player_pts))
        if computer_pts > player_pts:
            print('Computer dealer wins, You lost')
        elif computer_pts < player_pts:
            print('You win!! Computer dealer lost!!')
        else:
            print('Draw game')

    ## update the player's bank acct
    if player_bust == 1:
        player_bank.deduct_loss(bet_amount) 
        print('You have lost $'+str(bet_amount))
        print(player_bank)        
    elif computer_bust == 1:
        player_bank.add_winning(bet_amount)
        print('You have won $'+str(bet_amount))
        print(player_bank)
    elif computer_pts > player_pts:
        player_bank.deduct_loss(bet_amount) 
        print('You have lost $'+str(bet_amount))
        print(player_bank)
    elif computer_pts < player_pts:
        player_bank.add_winning(bet_amount)
        print('You have won $'+str(bet_amount))
        print(player_bank)
    else:
        print(player_bank)   
    
    ## ask player if he/she still want to play
    continue_or_not = player_continue()
    if continue_or_not == 'Y':
        game_on = True
    else:
        game_on = False 
    
### Show the final player's bank acct
print(player_bank)

if player_bank.account == 0:
    print('You are bankrupt')
```
